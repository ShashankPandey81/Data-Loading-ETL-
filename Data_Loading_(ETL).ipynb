Q9. Loading Strategy Selection
a) Should a Full Load or Incremental Load be used?

Incremental Load should be used.

b) Justify your choice.

Since this dataset represents daily sales data, only the new records generated each day need to be added to the database. Using an incremental load saves time, reduces system workload, and avoids loading the same records repeatedly. It is also more efficient than performing a full load every day, especially when the dataset becomes large.
